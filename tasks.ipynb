```python
import requests  
import json

url = "https://consumerbff.milkbasket.com/graphql"
payload = "{\"operationName\":\"fetchProductCategories\",\"variables\":{},\"query\":\"query fetchProductCategories {\\n  getProductCategories {\\n    id\\n    name\\n    img\\n    oos\\n    subCategories {\\n      id\\n      categoryId\\n      name\\n      img\\n      oos\\n      filters\\n      __typename\\n    }\\n    __typename\\n  }\\n}\"}"
headers = {
    "accept": "*/*",
    "authorization": "",
    "content-type": "application/json; charset=utf-8",
    "host": "consumerbff.milkbasket.com",
    "connection": "Keep-Alive",
    "accept-encoding": "gzip",
    "user-agent": "okhttp/4.9.2"
}
response = requests.post(url, data=payload, headers=headers).json()
```

```python
temp=response['data']['getProductCategories']
temp
```

```python
product_ids = []
product_sub_ids = []
for x in temp:
    product_ids.append(x['id'])
    product_cat = []
    for y in x['subCategories']:
        product_cat.append(y['id'])
    product_sub_ids.append(product_cat)
        

print(product_ids)
print(product_sub_ids)
```
